# Line Plots

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(10)
y1=x**2
y2=2*x+3

print(x)
print(y1)
print(y2)

# for ploting the graph

plt.plot(x,y1,color='red',label="Apple",marker='o')
plt.plot(x,y2,color='green',label="kiwi",linestyle='dashed',marker="*")
plt.style.use('seaborn')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over Time")
plt.legend()
plt.show()
```

```python
prices=np.array([1,2,3,4])**3
print(prices)
plt.plot(prices)# if we give only one array this prices become y axis and the indixes become x axis( these are [1,2,3,4] ) and power become y axis
plt.show()
```

# Scatter plots

```python
# Adjust the Size of plot
plt.figure(figsize=(5,5)) # it is basically 2D plot
plt.scatter(x,y1,color="red",label="Apple",marker="o")
plt.scatter(x,y2,color="green",label="Kiwi",marker="*",linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Prices")
plt.title("Scatter plots-->prices over time")
plt.legend()
plt.show()
```

```python
subjects=["Math","Physics","Chemistry","English"]
weightage=[30,25,25,20]
plt.pie(weightage,labels=subjects,explode=(0,0.1,0,0))
plt.show()
```

```python
subjects=["Math","Physics","Chemistry","English"]
weightage=[30,25,25,20]
plt.pie(weightage,labels=subjects,explode=(0.25,0.1,0,0.011))
plt.show()
```

```python
languages=["C","C++","JAVA","PYTHON","PHP"]  #Autopct-->string, used to label the wedges with their numeric value. The label will be placed inside the wedge. The format string will be fmt%pct.
student=[23,30,40,32,20]                    #  X--->array-like. The wedge sizes.
plt.pie(student,labels=languages,autopct="%1.2f%%")           # labels--->	list. A sequence of strings providing the labels for each wedge.
plt.show()
```

```python
subjects=["Math","Physics","Chemistry","English"]
weightage=[30,25,25,20]
plt.pie(weightage,labels=subjects,explode=(0.25,0.1,0,0.011),shadow="dark")

plt.show()
```

```python
plt.style.use('seaborn')
subjects=["Math","Physics","Chemistry","English"]
weightage=[30,25,25,20]
plt.pie(weightage,labels=subjects,explode=(0.25,0.1,0,0.011),shadow="dark",autopct='%1.1f%%')
plt.show()
```

# Histogram

```python
#xsn=np.random.random?
```

```python
xsn=np.random.random(100)
Sigma=3
u=55
X=np.round(xsn*Sigma+u)
X2=np.round(xsn*4+53)
print(X,X2)
```

```python
plt.style.use("seaborn")
plt.xlabel("Marks Range")
plt.ylabel("Prob/Freq of Students")
plt.title("Histogram")
plt.hist(X,alpha=0.8,label="Math")
plt.hist(X2,alpha=0.9,label="Pyhsics")
plt.legend()
plt.show()
```
