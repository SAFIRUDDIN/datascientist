# Histogram

```python
# Generate a normal Distribution (1-D)
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
u=5
Sigma=2

# Such a distribution is also called the Standard Normal Distribution
vals=u +Sigma*np.random.randn(1000)
print(vals.shape)
# print(vals)
plt.hist(vals,100)  # 100 is denoted number of bugets
plt.show()
```

```python
vals=np.round(vals)
z=np.unique(vals,return_counts=True)
print(z)
```

# Multivariate normal ditributtion

```python
# Bivariate
mean=np.array([0.0,0.0])
cov=np.array([[1,0.8],[0.8,1]])

mean2=np.array([5.0,6.0])
cov2=np.array([[1.3,0.2],[0.2,1.]])


dist=np.random.multivariate_normal(mean,cov,500)
dist2=np.random.multivariate_normal(mean2,cov2,400)
plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
print(dist.shape)
```
