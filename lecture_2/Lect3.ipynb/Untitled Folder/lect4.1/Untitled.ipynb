```python

import random
#random.choice()

import requests
```

```python
url=r"https://cb.lk/speech"
```

```python
result=requests.get(url)
```

```python
#rxsult.status_code
text= result.text
```

```python
print(text[:500])
```

```python
k=5
words={}
for i in range(k, len(text)):
    cut=text[i-k:i]
    next_char=text[i]
    
    
    if cut not in words:
        words[cut]={}
        
    if next_char not in words[cut]:
        words[cut][next_char]=0
        
    words[cut][next_char]+=1
words
```

```python
seed

=text[400:500]









```

```python
for i in range(500):
    cut=seed[-k:]
    sorted_words=sorted(words[cut].items(),key=lambda item:-item[1])
    top_three=sorted_words[:3]
    best_chr=random.choice(top_three)[0]
    seed=seed+best_chr
```
